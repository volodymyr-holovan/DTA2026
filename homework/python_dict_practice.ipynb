# Розділ 1. Словники (`dict`)

**Швидке нагадування:**

Словник — це колекція пар `ключ : значення`, створюється у фігурних дужках:
```python
student = {"name": "Олена", "age": 20, "city": "Київ"}
```

**Основні операції:**
- Доступ: `student["name"]`
- Додавання/зміна: `student["email"] = "olena@example.com"`
- Видалення: `del student["age"]`
- Перевірка: `if "name" in student:`
- Безпечне отримання: `student.get("phone")` — повертає `None`, якщо ключа немає
- Довжина: `len(student)` — кількість пар

**Перебір у циклі:**
- `for key in d.keys():` — перебір ключів
- `for value in d.values():` — перебір значень
- `for key, value in d.items():` — перебір пар

### Завдання 20. Інвентар магазину ⭐⭐⭐

Дано словник з товарами магазину, де значення — це інший словник з ціною і кількістю на складі. Виконайте:
1. Виведіть загальну вартість всього товару на складі (ціна × кількість для кожного).
2. Виведіть назви товарів, яких залишилось менше 20 штук.
3. Виведіть найдорожчий товар.

````python
inventory = {
    "Ноутбук":   {"price": 18500, "stock": 12},
    "Телефон":   {"price": 9200,  "stock": 35},
    "Планшет":   {"price": 7800,  "stock": 8},
    "Навушники": {"price": 850,   "stock": 120},
    "Клавіатура":{"price": 1200,  "stock": 45},
    "Мишка":     {"price": 450,   "stock": 18}
}
# Ваш код:
total_value = 0
low_stock = []
max_item = ""
max_price = 0

for name, data in inventory.items():
    total_value += data["price"] * data["stock"]

    if data["stock"] < 20:
        low_stock.append(name)

    if data["price"] > max_price:
        max_price = data["price"]
        max_item = name

print("Загальна вартість:", total_value)
print("Мало на складі:", low_stock)
print("Найдорожчий товар:", max_item, max_price)
````

> 💡 **Підказка:** Створіть порожній словник `revenue = {}`. У циклі для кожної покупки: якщо її категорія вже є у словнику — додайте ціну до існуючого значення, якщо немає — створіть запис з ціною як початковим значенням.

Створіть словник-телефонну книгу з 5 контактами. Напишіть програму, яка у циклі `while`:
1. Запитує у користувача дію: `пошук`, `додати`, `видалити`, `вихід`
2. Залежно від вибору — виконує відповідну дію
3. Завершується, коли користувач введе `вихід`

> 💡 **Підказка:** Використайте `while True:` (нескінченний цикл) і умову завершення з `break`, АБО змінну-прапорець, наприклад `running = True`.

````python
# Ваш код для завдання 22:
while True:
    print("\nДії: пошук / додати / видалити / вихід")
    action = input("Введіть дію: ").lower()

    if action == "пошук":
        name = input("Ім'я: ")
        if name in phonebook:
            print(phonebook[name])
        else:
            print("Контакт не знайдено")

    elif action == "додати":
        name = input("Ім'я: ")
        number = input("Телефон: ")
        phonebook[name] = number
        print("Додано")

    elif action == "видалити":
        name = input("Ім'я: ")
        if name in phonebook:
            del phonebook[name]
            print("Видалено")
        else:
            print("Немає такого контакту")

    elif action == "вихід":
        break

    else:
        print("Невідома дія")
````

### Завдання 23. Морфологічний словник ⭐⭐⭐

Дано список слів. Створіть словник, де ключ — це довжина слова, а значення — список слів цієї довжини.

**Приклад:** для `["кіт", "пес", "слон", "лев"]` → `{3: ["кіт", "пес", "лев"], 4: ["слон"]}`

> 💡 **Підказка:** Перевіряйте, чи є ключ-довжина у словнику. Якщо є — додайте слово до існуючого списку через `.append()`. Якщо немає — створіть новий ключ зі списком, у якому одне слово.

````python
words = ["кіт", "програма", "сонце", "пес", "комп'ютер", "стіл", "лев", "ніч", "школа", "день"]
# Ваш код:
length_dict = {}

for word in words:
    length = len(word)

    if length in length_dict:
        length_dict[length].append(word)
    else:
        length_dict[length] = [word]

print(length_dict)
````

### Завдання 24. Найпопулярніший товар ⭐⭐⭐

Дано список покупок (кожна — словник з полями `customer` і `product`). Знайдіть найпопулярніший товар (той, що купляли найбільше разів) і виведіть його назву та кількість покупок.

> 💡 **Підказка:** Спочатку зробіть словник `{товар: кількість}`. Потім у ньому знайдіть товар з максимальною кількістю.

````python
orders = [
    {"customer": "Олена",   "product": "Кава"},
    {"customer": "Андрій",  "product": "Чай"},
    {"customer": "Марія",   "product": "Кава"},
    {"customer": "Олена",   "product": "Тістечко"},
    {"customer": "Богдан",  "product": "Кава"},
    {"customer": "Софія",   "product": "Чай"},
    {"customer": "Петро",   "product": "Кава"},
    {"customer": "Дарʼя",   "product": "Тістечко"},
    {"customer": "Михайло", "product": "Кава"},
    {"customer": "Олена",   "product": "Кава"}
]
# Ваш код:
products = {}

for order in orders:
    product = order["product"]

    if product in products:
        products[product] += 1
    else:
        products[product] = 1

most_popular = max(products, key=products.get)

print("Найпопулярніший товар:", most_popular)
print("Кількість покупок:", products[most_popular])
````

````python
attendance = {
    "Олена":   [1, 1, 1, 1, 1],
    "Андрій":  [1, 0, 1, 1, 0],
    "Марія":   [1, 1, 1, 1, 1],
    "Богдан":  [0, 1, 0, 1, 1],
    "Софія":   [1, 1, 1, 0, 1],
    "Петро":   [0, 0, 1, 1, 1],
    "Дарʼя":   [1, 1, 1, 1, 0],
    "Михайло": [1, 0, 0, 1, 1]
}
# Ваш код:
days = ["Понеділок", "Вівторок", "Середа", "Четвер", "Пʼятниця"]

# Скільки днів кожен студент був присутній
for student, visits in attendance.items():
    present_days = sum(visits)
    print(student, "-", present_days, "днів")

print()

# Хто був присутній усі 5 днів
print("Присутні всі 5 днів:")

for student, visits in attendance.items():
    if sum(visits) == 5:
        print(student)

print()

# У який день найбільше прогулів
absences = [0, 0, 0, 0, 0]

for visits in attendance.values():
    for i in range(5):
        if visits[i] == 0:
            absences[i] += 1

max_absence_day = absences.index(max(absences))

print("Найбільше прогулів у день:", days[max_absence_day])

# Загальний відсоток відвідуваності
total_present = 0
total_days = 0

for visits in attendance.values():
    total_present += sum(visits)
    total_days += len(visits)

percentage = total_present / total_days * 100

print("Загальний відсоток відвідуваності:", round(percentage, 2), "%")
````

Дано список матчів. Кожен матч — це словник `{"team_a": ..., "team_b": ..., "score_a": ..., "score_b": ...}`. Створіть таблицю очок: за перемогу — 3 очка, за нічию — 1, за поразку — 0. Виведіть команди у порядку від найбільшої до найменшої кількості очок.

````python
matches = [
    {"team_a": "Динамо",     "team_b": "Шахтар",     "score_a": 1, "score_b": 2},
    {"team_a": "Зоря",       "team_b": "Динамо",     "score_a": 0, "score_b": 1},
    {"team_a": "Шахтар",     "team_b": "Дніпро",     "score_a": 3, "score_b": 1},
    {"team_a": "Динамо",     "team_b": "Ворскла",    "score_a": 2, "score_b": 2},
    {"team_a": "Шахтар",     "team_b": "Зоря",       "score_a": 4, "score_b": 0},
    {"team_a": "Дніпро",     "team_b": "Ворскла",    "score_a": 1, "score_b": 1},
    {"team_a": "Зоря",       "team_b": "Дніпро",     "score_a": 2, "score_b": 1},
    {"team_a": "Ворскла",    "team_b": "Шахтар",     "score_a": 0, "score_b": 3}
]
# Ваш код:
points = {}

for match in matches:
    team_a = match["team_a"]
    team_b = match["team_b"]
    score_a = match["score_a"]
    score_b = match["score_b"]

    if team_a not in points:
        points[team_a] = 0

    if team_b not in points:
        points[team_b] = 0

    if score_a > score_b:
        points[team_a] += 3
    elif score_b > score_a:
        points[team_b] += 3
    else:
        points[team_a] += 1
        points[team_b] += 1

sorted_teams = sorted(points.items(), key=lambda x: x[1], reverse=True)

print("Таблиця очок:")

for team, pts in sorted_teams:
    print(team, "-", pts)
````

3. Найпопулярніший тип тренування
4. Середня тривалість одного тренування

````python
workouts = [
    {"date": "2024-05-01", "type": "біг",       "duration_min": 30},
    {"date": "2024-05-02", "type": "йога",      "duration_min": 45},
    {"date": "2024-05-03", "type": "плавання",  "duration_min": 60},
    {"date": "2024-05-04", "type": "біг",       "duration_min": 40},
    {"date": "2024-05-06", "type": "йога",      "duration_min": 50},
    {"date": "2024-05-07", "type": "біг",       "duration_min": 35},
    {"date": "2024-05-08", "type": "плавання",  "duration_min": 45},
    {"date": "2024-05-10", "type": "йога",      "duration_min": 60},
    {"date": "2024-05-11", "type": "біг",       "duration_min": 50}
]
# Ваш код:
total_minutes = 0
types = {}
count_types = {}

for workout in workouts:
    duration = workout["duration_min"]
    workout_type = workout["type"]

    total_minutes += duration

    if workout_type in types:
        types[workout_type] += duration
        count_types[workout_type] += 1
    else:
        types[workout_type] = duration
        count_types[workout_type] = 1

print("Загальний час:", round(total_minutes / 60, 2), "год")

print("\nЧас по типах:")
for t, mins in types.items():
    print(t, "-", mins, "хв")

popular = max(count_types, key=count_types.get)

print("\nНайпопулярніший тип:", popular)

average = total_minutes / len(workouts)

print("Середня тривалість:", round(average, 2), "хв")
````

- `поповнити` — запитати суму і додати до балансу
- `історія` — показати список усіх операцій
- `вихід` — завершити

Зберігайте кожну операцію у списку як словник `{"type": "зняти", "amount": 500}`.

> 💡 **Підказка:** Створіть `history = []` перед циклом. Кожну операцію додавайте через `.append()`.

````python
# Ваш код для завдання 45:
balance = 5000
history = []

while True:
    action = input("\nВведіть дію: ").lower()

    if action == "баланс":
        print("Баланс:", balance)

    elif action == "зняти":
        amount = float(input("Сума: "))

        if amount <= balance:
            balance -= amount
            history.append({"type": "зняти", "amount": amount})
            print("Гроші знято")
        else:
            print("Недостатньо коштів")

    elif action == "поповнити":
        amount = float(input("Сума: "))
        balance += amount
        history.append({"type": "поповнити", "amount": amount})
        print("Рахунок поповнено")

    elif action == "історія":
        for item in history:
            print(item)

    elif action == "вихід":
        break

    else:
        print("Невідома команда")
````

### Завдання 46. Менеджер задач (todo) ⭐⭐⭐

Створіть програму-менеджер задач. Задачі зберігаються у списку словників: `{"task": "...", "done": False}`. У циклі `while` запитуйте дію:
- `додати` — додати нову задачу
- `показати` — показати всі задачі з номерами і статусом
- `виконати` — за номером позначити задачу як виконану
- `видалити` — за номером видалити задачу
- `вихід` — завершити

````python
# Ваш код для завдання 46:
tasks = []

while True:
    action = input("\nДія: ").lower()

    if action == "додати":
        task_text = input("Нова задача: ")

        tasks.append({
            "task": task_text,
            "done": False
        })

    elif action == "показати":
        for i, task in enumerate(tasks, start=1):
            status = "✓" if task["done"] else "✗"
            print(f"{i}. [{status}] {task['task']}")

    elif action == "виконати":
        number = int(input("Номер задачі: ")) - 1

        if 0 <= number < len(tasks):
            tasks[number]["done"] = True

    elif action == "видалити":
        number = int(input("Номер задачі: ")) - 1

        if 0 <= number < len(tasks):
            tasks.pop(number)

    elif action == "вихід":
        break

    else:
        print("Невідома команда")
````

````python
logs = [
    {"level": "INFO",    "message": "Сервер запущено",     "hour": 8},
    {"level": "INFO",    "message": "Користувач увійшов", "hour": 9},
    {"level": "WARNING", "message": "Повільний запит",     "hour": 9},
    {"level": "ERROR",   "message": "Помилка БД",          "hour": 10},
    {"level": "ERROR",   "message": "Тайм-аут запиту",     "hour": 10},
    {"level": "INFO",    "message": "Запит виконано",      "hour": 11},
    {"level": "WARNING", "message": "Низько місця на диску","hour": 12},
    {"level": "ERROR",   "message": "Помилка БД",          "hour": 14},
    {"level": "INFO",    "message": "Резервна копія",      "hour": 15},
    {"level": "ERROR",   "message": "Збій авторизації",    "hour": 14},
    {"level": "ERROR",   "message": "Помилка БД",          "hour": 14},
    {"level": "INFO",    "message": "Користувач вийшов",   "hour": 17}
]
# Ваш код:
levels = {}
errors_by_hour = {}
error_messages = []
hours = set()

for log in logs:
    level = log["level"]
    hour = log["hour"]

    hours.add(hour)

    if level in levels:
        levels[level] += 1
    else:
        levels[level] = 1

    if level == "ERROR":
        error_messages.append(log["message"])

        if hour in errors_by_hour:
            errors_by_hour[hour] += 1
        else:
            errors_by_hour[hour] = 1

print("Кількість записів:", levels)

max_error_hour = max(errors_by_hour, key=errors_by_hour.get)

print("Найбільше ERROR-ів о:", max_error_hour)

print("\nПовідомлення ERROR:")
for msg in error_messages:
    print(msg)

print("\nГодини без ERROR:")

for hour in hours:
    if hour not in errors_by_hour:
        print(hour)
````

### Завдання 54. Магазин з фільтрами ⭐⭐⭐

Дано список товарів — словники з полями `name`, `category`, `price`, `rating`, `in_stock`. У циклі `while` запитуйте у користувача фільтр:
- `категорія` — запитайте назву категорії, виведіть товари цієї категорії
- `ціна` — запитайте максимальну ціну, виведіть товари дешевше
- `рейтинг` — виведіть товари з рейтингом ≥ 4.5
- `в наявності` — виведіть товари, у яких `in_stock = True`
- `топ` — виведіть 3 найдорожчі товари
- `вихід` — завершити

````python
products = [
    {"name": "Ноутбук HP",     "category": "Електроніка", "price": 22000, "rating": 4.6, "in_stock": True},
    {"name": "Ноутбук Lenovo", "category": "Електроніка", "price": 18500, "rating": 4.3, "in_stock": True},
    {"name": "Телефон Samsung","category": "Електроніка", "price": 12000, "rating": 4.7, "in_stock": False},
    {"name": "Книга Python",   "category": "Книги",       "price": 450,   "rating": 4.8, "in_stock": True},
    {"name": "Книга SQL",      "category": "Книги",       "price": 380,   "rating": 4.2, "in_stock": True},
    {"name": "Кросівки Nike",  "category": "Одяг",        "price": 2800,  "rating": 4.5, "in_stock": True},
    {"name": "Куртка Adidas",  "category": "Одяг",        "price": 3500,  "rating": 4.0, "in_stock": False},
    {"name": "Навушники",      "category": "Електроніка", "price": 1200,  "rating": 4.6, "in_stock": True},
    {"name": "Чайник",         "category": "Побутова",    "price": 850,   "rating": 4.4, "in_stock": True},
    {"name": "Блендер",        "category": "Побутова",    "price": 1500,  "rating": 4.7, "in_stock": False}
]
# Ваш код:
while True:
    action = input("\nФільтр: ").lower()

    if action == "категорія":
        category = input("Назва категорії: ")

        for product in products:
            if product["category"].lower() == category.lower():
                print(product["name"])

    elif action == "ціна":
        max_price = float(input("Максимальна ціна: "))

        for product in products:
            if product["price"] <= max_price:
                print(product["name"])

    elif action == "рейтинг":
        for product in products:
            if product["rating"] >= 4.5:
                print(product["name"])

    elif action == "в наявності":
        for product in products:
            if product["in_stock"]:
                print(product["name"])

    elif action == "топ":
        sorted_products = sorted(products, key=lambda x: x["price"], reverse=True)

        for product in sorted_products[:3]:
            print(product["name"], "-", product["price"])

    elif action == "вихід":
        break

    else:
        print("Невідомий фільтр")
````
