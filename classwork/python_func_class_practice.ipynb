Голосні: `а е и і о у я ю є ї`.

### Завдання 10. Перевірка простого числа ⭐⭐

Створіть функцію `is_prime(n)`, яка повертає `True`, якщо число просте, і `False` інакше.

**Нагадування:** просте число ділиться лише на 1 і на саме себе. Числа менші за 2 — не прості.

**Перевірте:**
```python
print(is_prime(7))    # True
print(is_prime(10))   # False
print(is_prime(1))    # False
```

> 💡 **Підказка:** У циклі від 2 до n-1 перевіряйте, чи ділиться `n` на якесь число без остачі. Якщо знайшли дільник — повертайте одразу `False`. Якщо цикл завершився без знайдених дільників — `True`.

````python
# Ваш код для завдання 10:
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

print(is_prime(7))
print(is_prime(10))
print(is_prime(1))
````

### Завдання 11. Факторіал ⭐⭐

Створіть функцію `factorial(n)`, яка обчислює факторіал числа `n`.

**Нагадування:** `n! = 1 * 2 * 3 * ... * n`. Наприклад, `5! = 120`.

**Перевірте:**
```python

print(factorial(0))   # 1
```

````python
# Ваш код для завдання 11:
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print(factorial(5))
print(factorial(0))
````


````python
# Ваш код для завдання 12:
# Спосіб 1 — через зріз
def reverse_string(text):
    return text[::-1]

# Спосіб 2 — через цикл
def reverse_string_loop(text):
    result = ""
    for char in text:
        result = char + result
    return result

print(reverse_string("Python"))
print(reverse_string_loop("Python"))
````

### Завдання 13. Середнє арифметичне ⭐⭐

Створіть функцію `average(numbers)`, яка приймає список чисел і повертає їхнє середнє арифметичне. Якщо список порожній — поверніть 0.

> 💡 **Підказка:** Не забудьте перевірити порожній список перед діленням, щоб уникнути ділення на нуль.

````python
# Ваш код для завдання 13:
def average(numbers):
    if len(numbers) == 0:
        return 0
    total = 0
    for num in numbers:
        total += num
    return total / len(numbers)

print(average([1, 2, 3, 4, 5]))
print(average([]))
````

### Завдання 14. Підрахунок цифр у числі ⭐⭐⭐

Створіть функцію `count_digits(n)`, яка приймає ціле число і повертає кількість цифр у ньому.

**Перевірте:**
```python
print(count_digits(12345))   # 5
print(count_digits(7))       # 1
print(count_digits(0))       # 1
```

Реалізуйте через **математику** (без перетворення в рядок).

> 💡 **Підказка:** У циклі `while n > 0` ділимо число націло на 10 (`n = n // 10`) і збільшуємо лічильник. Окремо обробіть випадок `n == 0`.

````python
# Ваш код для завдання 14:
def count_digits(n):
    if n == 0:
        return 1
    n = abs(n)  # на випадок від'ємного числа
    count = 0
    while n > 0:
        n = n // 10
        count += 1
    return count

print(count_digits(12345))  # 5
print(count_digits(7))      # 1
print(count_digits(0))      # 1
````

### Завдання 16. Найдовше слово ⭐⭐⭐

Створіть функцію `longest_word(sentence)`, яка приймає рядок-речення і повертає найдовше слово.

**Перевірте:**
```python
print(longest_word("Python це чудова мова програмування"))   # "програмування"
```

> 💡 **Підказка:** Розбийте речення на список слів через `.split()`. Потім переберіть і знайдіть найдовше, порівнюючи через `len()`.

````python
# Ваш код для завдання 16:
def longest_word(sentence):
    words = sentence.split()
    longest = ""
    for word in words:
        if len(word) > len(longest):
            longest = word
    return longest

print(longest_word("Python це чудова мова програмування"))
````

### Завдання 17. Калькулятор з операцією ⭐⭐⭐

Створіть функцію `calculate(a, b, operation)`, яка виконує одну з операцій залежно від рядка `operation`:
- `"+"` — додавання
- `"-"` — віднімання
- `"*"` — множення
- `"/"` — ділення

Якщо операція невідома — поверніть рядок `"Невідома операція"`. Якщо ділення на нуль — поверніть `"Помилка: ділення на нуль"`.

````python
# Ваш код для завдання 17:
def calculate(a, b, operation):
    if operation == "+":
        return a + b
    elif operation == "-":
        return a - b
    elif operation == "*":
        return a * b
    elif operation == "/":
        if b == 0:
            return "Помилка: ділення на нуль"
        return a / b
    else:
        return "Невідома операція"

print(calculate(10, 5, "+"))
print(calculate(10, 5, "-"))
print(calculate(10, 5, "*"))
print(calculate(10, 0, "/"))
print(calculate(10, 5, "^"))
````

### Завдання 18. Числа Фібоначчі ⭐⭐⭐

Створіть функцію `fibonacci(n)`, яка повертає **список** з перших `n` чисел Фібоначчі.

**Нагадування:** кожне число — сума двох попередніх. Перші два: 0 і 1.

**Перевірте:**
```python
print(fibonacci(7))   # [0, 1, 1, 2, 3, 5, 8]
```

> 💡 **Підказка:** Створіть список з перших двох чисел `[0, 1]`. У циклі додавайте суму двох останніх елементів через `.append()`.

````python
# Ваш код для завдання 18:
def fibonacci(n):
    if n <= 0:
        return []
    if n == 1:
        return [0]
    fib = [0, 1]
    for i in range(2, n):
        fib.append(fib[-1] + fib[-2])
    return fib

print(fibonacci(7))  # [0, 1, 1, 2, 3, 5, 8]
````

### Завдання 20. Клас Rectangle ⭐

Створіть клас `Rectangle` з атрибутами `width` і `height`. Додайте два методи:

2. `perimeter()` — повертає периметр

Створіть прямокутник 5x3 і виведіть його площу та периметр.

````python
# Ваш код для завдання 20:
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

rect = Rectangle(5, 3)
print(rect.area())
print(rect.perimeter())
````

### Завдання 22. Клас Circle ⭐

Створіть клас `Circle` з атрибутом `radius`. Додайте методи:
1. `area()` — площа кола (`π * r²`, використайте `3.14` або `math.pi`)
2. `circumference()` — довжина кола (`2 * π * r`)

````python
# Ваш код для завдання 22:
import math

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

    def circumference(self):
        return 2 * math.pi * self.radius

c = Circle(5)
print(f"Площа: {c.area():.2f}")          # 78.54
print(f"Довжина кола: {c.circumference():.2f}")  # 31.42
````

### Завдання 23. Клас Student ⭐⭐

Створіть клас `Student` з атрибутами `name` і порожнім списком `grades`. Додайте методи:
1. `add_grade(grade)` — додає оцінку до списку
2. `average()` — повертає середній бал (0, якщо оцінок ще немає)
3. `__str__()` — повертає рядок виду `"Студент <name>, середній бал: <average>"`

Створіть студента, додайте 4 оцінки, виведіть об'єкт через `print()`.

````python
# Ваш код для завдання 23:
class Student:
    def __init__(self, name):
        self.name = name
        self.grades = []

    def add_grade(self, grade):
        self.grades.append(grade)

    def average(self):
        if len(self.grades) == 0:
            return 0
        return sum(self.grades) / len(self.grades)

    def __str__(self):
        return f"Студент {self.name}, середній бал: {self.average():.1f}"

student = Student("Олена")
student.add_grade(90)
student.add_grade(85)
student.add_grade(92)
student.add_grade(88)
print(student)
````

### Завдання 24. Клас Counter ⭐⭐

Створіть клас `Counter` (лічильник). У конструкторі задайте `self.value = 0`. Додайте методи:
1. `increment()` — збільшити на 1
2. `decrement()` — зменшити на 1
3. `reset()` — скинути в 0
4. `__str__()` — повернути поточне значення як рядок

Створіть лічильник, виконайте 5 increment і 2 decrement, виведіть результат.

````python
# Ваш код для завдання 24:
class Counter:
    def __init__(self):
        self.value = 0

    def increment(self):
        self.value += 1

    def decrement(self):
        self.value -= 1

    def reset(self):
        self.value = 0

    def __str__(self):
        return str(self.value)

counter = Counter()
for _ in range(5):
    counter.increment()
counter.decrement()
counter.decrement()
print(counter)
````

````python
# Ваш код для завдання 33:
class Student:
    def __init__(self, name):
        self.name = name
        self.grades = []

    def add_grade(self, grade):
        self.grades.append(grade)

    def average(self):
        if not self.grades:
            return 0
        return sum(self.grades) / len(self.grades)


def top_student(students):
    return max(students, key=lambda s: s.average())

def group_average(students):
    if not students:
        return 0
    return sum(s.average() for s in students) / len(students)

def print_report(students):
    print(f"{'Ім\'я':<15} {'Середній бал':>12}")
    print("-" * 28)
    for s in students:
        print(f"{s.name:<15} {s.average():>12.1f}")
    print("-" * 28)
    print(f"{'Загальний середній:':<15} {group_average(students):>12.1f}")


# Тест
s1 = Student("Олена");   [s1.add_grade(g) for g in [90, 85, 92, 88]]
s2 = Student("Максим");  [s2.add_grade(g) for g in [75, 80, 70, 78]]
s3 = Student("Софія");   [s3.add_grade(g) for g in [95, 98, 92, 97]]
s4 = Student("Андрій");  [s4.add_grade(g) for g in [60, 72, 65, 68]]
s5 = Student("Ірина");   [s5.add_grade(g) for g in [85, 88, 82, 90]]

students = [s1, s2, s3, s4, s5]
print_report(students)
print(f"\nТоп-студент: {top_student(students).name}")
````

2. `bestseller(products)` — товар з найбільшою кількістю продажів
3. `revenue_by_category(products)` — повертає **словник** `{категорія: виручка}`
4. `products_above_price(products, min_price)` — список товарів дорожче за `min_price`

````python
# Ваш код для завдання 34:
class Product:
    def __init__(self, name, price, category, quantity_sold):
        self.name = name
        self.price = price
        self.category = category
        self.quantity_sold = quantity_sold


def total_revenue(products):
    return sum(p.price * p.quantity_sold for p in products)

def bestseller(products):
    return max(products, key=lambda p: p.quantity_sold)

def revenue_by_category(products):
    result = {}
    for p in products:
        revenue = p.price * p.quantity_sold
        result[p.category] = result.get(p.category, 0) + revenue
    return result

def products_above_price(products, min_price):
    return [p for p in products if p.price > min_price]


# Тест
products = [
    Product("Молоко", 45, "Молочне", 200),
    Product("Сир", 180, "Молочне", 80),
    Product("Хліб", 30, "Хлібобулочне", 350),
    Product("Булочка", 15, "Хлібобулочне", 120),
    Product("Кава", 250, "Напої", 60),
    Product("Сік", 55, "Напої", 150),
    Product("Яблука", 40, "Фрукти", 180),
    Product("Банани", 35, "Фрукти", 220),
]

print(f"Загальна виручка: {total_revenue(products)} грн")
print(f"Бестселер: {bestseller(products).name}")
print("\nВиручка по категоріях:")
for category, revenue in revenue_by_category(products).items():
    print(f"  {category}: {revenue} грн")
print("\nТовари дорожче 50 грн:")
for p in products_above_price(products, 50):
    print(f"  {p.name} — {p.price} грн")
````

3. `total_matches(teams)` — загальна кількість зіграних матчів (сума всіх перемог + нічиїх + поразок, поділена на 2, бо кожен матч рахується двічі)

````python
# Ваш код для завдання 36:
class Team:
    def __init__(self, name, wins, draws, losses):
        self.name = name
        self.wins = wins
        self.draws = draws
        self.losses = losses

    def points(self):
        return self.wins * 3 + self.draws


def tournament_table(teams):
    sorted_teams = sorted(teams, key=lambda t: t.points(), reverse=True)
    print(f"{'#':<3} {'Команда':<15} {'В':>4} {'Н':>4} {'П':>4} {'Очки':>6}")
    print("-" * 38)
    for i, t in enumerate(sorted_teams, 1):
        print(f"{i:<3} {t.name:<15} {t.wins:>4} {t.draws:>4} {t.losses:>4} {t.points():>6}")

def champion(teams):
    return max(teams, key=lambda t: t.points())

def total_matches(teams):
    return sum(t.wins + t.draws + t.losses for t in teams) // 2


# Тест
teams = [
    Team("Динамо",     8, 4, 2),
    Team("Шахтар",     9, 2, 3),
    Team("Металіст",   6, 3, 5),
    Team("Дніпро",     5, 5, 4),
    Team("Карпати",    3, 4, 7),
    Team("Чорноморець",2, 2, 10),
]

tournament_table(teams)
print(f"\nЧемпіон: {champion(teams).name}")
print(f"Зіграно матчів: {total_matches(teams)}")
````

### Завдання 37. Банківська система ⭐⭐⭐

Створіть клас `BankAccount` з полями `owner`, `balance`, методами `deposit(amount)`, `withdraw(amount)`, `__str__()`.

Створіть **функції** для роботи з кількома рахунками:
1. `transfer(from_account, to_account, amount)` — переказує гроші з одного рахунку на інший (перевірте, чи вистачає коштів)
2. `total_money(accounts)` — загальна сума на всіх рахунках
3. `richest_client(accounts)` — рахунок з найбільшим балансом
4. `accounts_above(accounts, min_balance)` — список рахунків з балансом > `min_balance`

Створіть 4-5 рахунків і протестуйте все.

> 💡 **Підказка:** У `transfer` спочатку зробіть `withdraw` з першого рахунку. Якщо вийшло (баланс зменшився) — зробіть `deposit` на другий. Можете додати метод `withdraw()`, який повертає `True`/`False` залежно від успіху.

````python
# Ваш код для завдання 37:
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount > self.balance:
            print(f"Недостатньо коштів у {self.owner} (є {self.balance} грн)")
            return False
        self.balance -= amount
        return True

    def __str__(self):
        return f"{self.owner}: {self.balance} грн"


def transfer(from_account, to_account, amount):
    if from_account.withdraw(amount):
        to_account.deposit(amount)
        print(f"Переказ {amount} грн: {from_account.owner} → {to_account.owner}")

def total_money(accounts):
    return sum(a.balance for a in accounts)

def richest_client(accounts):
    return max(accounts, key=lambda a: a.balance)

def accounts_above(accounts, min_balance):
    return [a for a in accounts if a.balance > min_balance]


# Тест
a1 = BankAccount("Олена", 5000)
a2 = BankAccount("Максим", 12000)
a3 = BankAccount("Софія", 800)
a4 = BankAccount("Андрій", 3500)
a5 = BankAccount("Ірина", 9200)

accounts = [a1, a2, a3, a4, a5]

transfer(a2, a3, 2000)
transfer(a3, a1, 10000)  # не вистачить

print(f"\nЗагальна сума: {total_money(accounts)} грн")
print(f"Найбагатший: {richest_client(accounts).owner}")
print("\nРахунки понад 4000 грн:")
for a in accounts_above(accounts, 4000):
    print(f"  {a}")
````

### Завдання 38. Школа з кількома класами ⭐⭐⭐

Створіть клас `Student` (ім'я + оцінки). Створіть клас `SchoolClass` з полем `name` (наприклад, `"7-А"`) і списком студентів. Додайте до `SchoolClass` методи:
1. `add_student(student)` — додає студента
2. `class_average()` — середній бал по класу

Створіть **функції**, які приймають список об'єктів `SchoolClass`:
1. `best_class(classes)` — клас з найвищим середнім балом
2. `total_students(classes)` — загальна кількість учнів у школі
3. `school_report(classes)` — виводить звіт по школі (для кожного класу — назва, кількість учнів, середній бал)

Створіть 3 класи, по 3-4 студенти у кожному.

````python
# Ваш код для завдання 38:
class Student:
    def __init__(self, name):
        self.name = name
        self.grades = []

    def add_grade(self, grade):
        self.grades.append(grade)

    def average(self):
        if not self.grades:
            return 0
        return sum(self.grades) / len(self.grades)

class SchoolClass:
    def __init__(self, name):
        self.name = name
        self.students = []

    def add_student(self, student):
        self.students.append(student)

    def class_average(self):
        if not self.students:
            return 0
        return sum(s.average() for s in self.students) / len(self.students)


def best_class(classes):
    return max(classes, key=lambda c: c.class_average())

def total_students(classes):
    return sum(len(c.students) for c in classes)

def school_report(classes):
    print(f"{'Клас':<8} {'Учнів':>6} {'Середній бал':>13}")
    print("-" * 30)
    for c in classes:
        print(f"{c.name:<8} {len(c.students):>6} {c.class_average():>13.1f}")
    print(f"\nУсього учнів: {total_students(classes)}")
    print(f"Найкращий клас: {best_class(classes).name}")


# Тест
def make_student(name, grades):
    s = Student(name)
    for g in grades:
        s.add_grade(g)
    return s

c1 = SchoolClass("7-А")
c1.add_student(make_student("Олена",  [90, 88, 92]))
c1.add_student(make_student("Максим", [75, 80, 78]))
c1.add_student(make_student("Ірина",  [95, 92, 97]))
c1.add_student(make_student("Данило", [70, 65, 72]))

c2 = SchoolClass("7-Б")
c2.add_student(make_student("Софія",  [88, 90, 85]))
c2.add_student(make_student("Андрій", [60, 68, 65]))
c2.add_student(make_student("Аліна",  [82, 79, 84]))

c3 = SchoolClass("8-А")
c3.add_student(make_student("Віктор", [95, 98, 96]))
c3.add_student(make_student("Оксана", [91, 93, 89]))
c3.add_student(make_student("Богдан", [87, 84, 88]))
c3.add_student(make_student("Тетяна", [76, 80, 78]))

school_report([c1, c2, c3])
````
